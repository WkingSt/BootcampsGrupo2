```python
#Promedio edad de clubes
df = pd.read_csv('players_20.csv')

# Seleccionar los 3 clubes específicos (reemplaza 'Club1', 'Club2', 'Club3' con los nombres reales de los clubes)
club1 = df[df['club'] == 'FC Barcelona']
club2 = df[df['club'] == 'Manchester City']
club3 = df[df['club'] == 'Liverpool']

# Calcular el promedio de edad para cada club
promedio_edad_club1 = club1['age'].mean()
promedio_edad_club2 = club2['age'].mean()
promedio_edad_club3 = club3['age'].mean()

# Crear un gráfico de barras con seaborn
clubes = ['FC Barcelona', 'Manchester City', 'Liverpool']
promedios_edad = [promedio_edad_club2, promedio_edad_club2, promedio_edad_club3]

sns.barplot(x=clubes, y=promedios_edad, palette='viridis')
plt.xlabel('Clubes')
plt.ylabel('Promedio de Edad')
plt.title('Promedio de Edad de Clubes')
plt.show()
```

```python
#Promedio de estatura
df = pd.read_csv('players_20.csv')

# Seleccionar los 3 clubes específicos (reemplaza 'Club1', 'Club2', 'Club3' con los nombres reales de los clubes)
club1 = df[df['club'] == 'FC Barcelona']
club2 = df[df['club'] == 'Manchester City']
club3 = df[df['club'] == 'Liverpool']

# Calcular el promedio de edad para cada club
promedio_estatura_club1 = club1['height_cm'].mean()
promedio_estatura_club2 = club2['height_cm'].mean()
promedio_estatura_club3 = club3['height_cm'].mean()

# Crear un gráfico de barras con seaborn
clubes = ['FC Barcelona', 'Manchester City', 'Liverpool']
promedios_estatura = [promedio_estatura_club2, promedio_estatura_club2, promedio_estatura_club3]

sns.barplot(x=clubes, y=promedios_estatura, palette='viridis')
plt.xlabel('Clubes')
plt.ylabel('Promedio de Estatura')
plt.title('Promedio de Estatura de Jugadores por Club')
plt.show()
```

```python
#Jugadores con mayor puntaje general
df = pd.read_csv('players_20.csv')

# Ordenar el DataFrame por puntaje general de manera descendente
df_sorted = df.sort_values(by='overall', ascending=False)

# Seleccionar los primeros N jugadores (puedes ajustar N según tus necesidades)
top_players = df_sorted.head(10)

# Crear un gráfico de barras horizontales con seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='overall', y='short_name', data=top_players, palette='viridis')
plt.xlabel('Puntaje General')
plt.ylabel('Jugador')
plt.title('Top 10 Jugadores con Mayor Puntaje General en FIFA 20')
plt.show()
```

```python
# Jugadores con mayor valor en el mercado
df = pd.read_csv('players_20.csv')

# Ordenar el DataFrame por valor de mercado de manera descendente
df_sorted = df.sort_values(by='value_eur', ascending=False)

# Seleccionar los primeros N jugadores (puedes ajustar N según tus necesidades)
top_players = df_sorted.head(10)

# Crear un gráfico de barras horizontales con seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='value_eur', y='short_name', data=top_players, palette='viridis')
plt.xlabel('Valor en el Mercado (EUR)')
plt.ylabel('Jugador')
plt.title('Top 10 Jugadores con Mayor Valor en el Mercado en FIFA 20')
plt.show()
```
